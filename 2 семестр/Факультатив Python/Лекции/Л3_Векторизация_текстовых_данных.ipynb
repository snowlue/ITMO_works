## Словарные методы

Для задачи сентимент-анализа и анализа эмоций текстовые данные могут векторизоваться при помощи тональных словарей. **Тональный словарь** представляет из себя набор слов, напротив каждого стоит метка (вес) его сентимента или эмоций.

image.png

**Алгоритм векторизации текста при помощи тональных словарей**


1. Предобработать текст. Использовать лемматизацию.
2. Создать переменную vector типа list, куда будем записывать наш вектор.
3. Пройтись по каждому токену.
4. Если токен есть в тональном словаре, то добавить вес слова из тонального словаря в vector.
5. Если токена нет в тональном словаре, то в vector добавить 0.
6. В результате получим, что длина вектора совпадает с длиной текста.
7. Считаем статистические данные из полученного vector 
7.1 Кол-во положительных слов
7.2 Кол-во отрицаильных слов
7.3 Среднее значение сентимента всех слов
7.4 Максмально положительное значение вектора
7.5 Максимально отрицательное значение вектора
7.6 ...
8. Записываем статистические данные из п.7 в переменную static_vector типа list. ***static_vector*** - вектор каждого текста, векторизиванного при помощи тональных словарей.

# Мешок слов, Bag-of-Words, BoW

**Bag of Words, BoW, «Мешок слов»** - метод представления текста. Рассматривает текст в виде неупорядоченного набора слов и представляет в виде матрицы, где строки соответствуют отдельному тексту, а столбцы слову

## BoW на Python

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#записываем метод векторизации в переменную vectorizer
vectorizer = CountVectorizer()
```

```python
#наш корпус текстов задаем вручную
corpus = ['вася любить кот',
          'кот жить улица',
          'кот любить рыба']
```

```python
X = vectorizer.fit_transform(corpus).toarray()
X
```

# TF-IDF

**Term Frequency Inverse Document Frequency, TF-IDF** – статистическая метрика, используемая для оценки важности слова в контексте документа.

## TF-IDF на Python

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#записываем метод векторизации в переменную vectorizer
vectorizer = TfidfVectorizer()
```

```python
#наш корпус текстов задаем вручную
corpus = ['вася любить кот',
          'кот жить улица',
          'кот любить рыба']
```

```python
X = vectorizer.fit_transform(corpus).toarray()
print(X)
```
